```python
import numpy as np 
import pandas as pd

# Defining the three dataframes indicating the gold, silver, and bronze medal counts
# of different countries
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
vertical_stack=pd.concat([gold,silver,bronze],axis=0,ignore_index=True)
dfp=vertical_stack.groupby('Country',sort=True).sum()
dfp.sort_values(by=['Medals'],inplace=True,ascending=False)
dfp['Medals']=dfp['Medals'].astype('float64')
dfp['Medals'].dtype
print(dfp)
```
